```python
import tensorflow as tf
from tensorflow.keras import models, layers
import matplotlib.pyplot as plt
import PIL
```

```python
IMAGE_SIZE = 256
BATCH_SIZE = 32
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    'PlantVillage',
    shuffle=True,
    image_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE
)
```

```python
EPOCHS = 10
```

```python
def get_data_splits(dataset, train_split=0.8, test_split=0.1, val_split=0.1, shuffle=True, shuffle_size=10000):
    if shuffle:
        dataset = dataset.shuffle(shuffle_size, seed=13)
    train_ds = dataset.take(int(len(dataset)*train_split))
    #test_ds = dataset.skip(int(len(dataset)*train_split))
    val_ds = dataset.skip(int(len(dataset)*train_split)).take(int(len(dataset)*val_split))
    test_ds = dataset.skip(int(len(dataset)*train_split)).skip(int(len(dataset)*val_split))
    return train_ds, val_ds, test_ds
    
```

```python
train_ds, val_ds, test_ds = get_data_splits(dataset)
```

```python
len(train_ds), len(test_ds), len(val_ds)
```

```python
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds = test_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds = val_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
resize_rescale = tf.keras.Sequential([
    layers.experimental.preprocessing.Resizing(IMAGE_SIZE, IMAGE_SIZE),
    layers.experimental.preprocessing.Rescaling(1.0/255)
])
```

```python
augmentation = tf.keras.Sequential([
    layers.experimental.preprocessing.RandomFlip('horizontal_and_vertical'),
    layers.experimental.preprocessing.RandomRotation(0.2)
])
```

```python

CHANNELS = 3
input_shape = (BATCH_SIZE, IMAGE_SIZE, IMAGE_SIZE, CHANNELS)
model = tf.keras.Sequential([
    resize_rescale,
    augmentation,
    layers.Conv2D(32, (3,3), activation='relu', input_shape=input_shape),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(3, activation='softmax')
])
model.build(input_shape=input_shape)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
histroy = model.fit(train_ds, epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=1, validation_data=val_ds)
```

```python
model.evaluate(test_ds)
```

```python
import os
model_version = max([int(i) for i in os.listdir('../models/') + [0]]) +1
model.save(f'../models/{model_version}', save_format='h5')
```
